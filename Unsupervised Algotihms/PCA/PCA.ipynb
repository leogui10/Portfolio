# Principal Component Analysis

## 1. Importamos las librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
```

## 2. Importamos los datos

```python
digits = load_digits()
X = digits.data
y = digits.target
```

## 3. EDA

```python
# Vemos la data cruda

# Mostrar los primeros 5 registros de los datos crudos
print(digits.data[:5])
```

```python
#Vemos como viene la Data

#Cuantas imagenes mostraremos
num_images = 10

#Creamos la figura
fig, axes = plt.subplots(1, num_images, figsize=(10, 3), subplot_kw={'xticks': [], 'yticks': []})

#Mostramos las primeras num_imagenes
for i in range(num_images):
    ax = axes[i]
    ax.imshow(X[i].reshape(8, 8), cmap='gray')
    ax.set_title(y[i])



```

## 4. Aplicamos PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

## 5. Graficamos los resultados

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.colorbar(scatter, label='Dígitos')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA de dígitos escritos a mano')
plt.grid()
```
