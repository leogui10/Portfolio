## Creamos subconjuntos para ver mas caracteristicas

```python
df_class0 = df[df['target'] == 0]
df_class1 = df[df['target'] == 1]
df_class2 = df[df['target'] == 2]
```

```python
df_class2.head(20)
```

```python
df_class0.alcohol.plot(kind="hist")
plt.title("Distribución de alcohol")
plt.xlabel("Distribución del alcohol")
plt.ylabel("Frecuencia")
plt.xticks(rotation= 0)
plt.show()
```

```python
df_class1.alcohol.plot(kind="hist")
plt.title("Distribución de alcohol")
plt.xlabel("Distribución del alcohol")
plt.ylabel("Frecuencia")
plt.xticks(rotation= 0)
plt.show()
```

```python
df_class2.alcohol.plot(kind="hist")
plt.title("Distribución de alcohol")
plt.xlabel("Distribución del alcohol")
plt.ylabel("Frecuencia")
plt.xticks(rotation= 0)
plt.show()
```

```python
print("Asimetría ", df_class2['alcohol'].skew())
```

## Relación entre variables y detección de atipicos

```python
sns.catplot(x="target", y="proline", data=df, kind="box", aspect=1.5)
plt.title("Boxplot Target vs Proline")
```

```python
sns.catplot(x="target", y="flavanoids", data=df, kind="box", aspect=1.5)
plt.title("Boxplot Target vs Flavanoids")
```
